## Pregunta 1: DataFrames y listas en python
Usando PeriodicTable, crea una tabla con 20 elementos químicos y en las columnas considera: nombre, simbolo, masa_atomica, numero_atomico y valencias.

```python
!pip install rdkit
from rdkit import Chem
PeriodicTable = Chem.GetPeriodicTable()
```

```python
element_symbols = []

for i in range (1,21):
    element_symbols.append(PeriodicTable.GetElementSymbol(i))

print(element_symbols)
```

```python
masa_atomica = [PeriodicTable.GetAtomicWeight(x) for x in element_symbols]
numero_atomico = [PeriodicTable.GetAtomicNumber(x) for x in element_symbols]
nombre_elemento = [PeriodicTable.GetElementName(x) for x in range (1,21)]
valencia = [PeriodicTable.GetValenceList(x) for x in range (1,21)]
print(masa_atomica)
print(numero_atomico)
print(nombre_elemento)
print(valencia)

import pandas as pd
data = {"nombre": nombre_elemento, "simbolo": element_symbols,
        "masa atomica": masa_atomica,"numero atomico": numero_atomico, "Valencias": valencia}
df = pd.DataFrame(data)
df
```

# Pregunta 2: Configuración electrónica
Escribe un programa que lea el número atómico y me indique el número de electrones de valencia.

```python
z = int(input("ingrese el numero de protones o fin para terminar: "))
gas_noble= [2, 10, 18, 36, 54, 86]
def electrones_valencia(z):
  if z >= 1 and z <= 2:
    return z
  elif z >= 3 and z <= 10:
    return z - gas_noble[0]
  elif z >= 11 and z <= 18:
    return z - gas_noble[1]
  elif z >= 19 and z <= 20:
    return z - gas_noble[2]
  elif z >= 21 and z <= 30:
    return gas_noble[0], "es un metal de transicion del grupo D"
  elif z >= 31 and z <= 36:
    return z - (gas_noble[2] + 10)
  elif z >= 37 and z <= 38:
    return z - gas_noble[3]
  elif z >= 39 and z <= 48:
    return gas_noble[0], "es un metal de transicion del grupo D"
  elif z >= 49 and z <= 54:
    return z - (gas_noble[3] + 10)
  elif z >= 55 and z <= 56:
    return z - gas_noble[4]
  elif z >= 57 and z <= 80:
    return gas_noble[0], "es un metal de transicion del grupo D o f"
  elif z >= 81 and z <= 86:
    return z - (gas_noble[4] + 24)
  elif z >= 87 and z <= 88:
    return z - gas_noble[5]
  elif z >= 89 and z <= 112:
    return gas_noble[0], "es un metal de transicion del grupo D o f"
  elif z >= 113 and z <= 118:
    return z - (gas_noble[5] + 24)
  else:
    return "Numero atomico no valido"

print("el numero de electrones de valencia segun el numero atomico", z, "es:", electrones_valencia(z))
```

```python
#Con los coeficientes se escribe la formula de la ecuacion
def get_sub(x):
  normal = "0123456789"
  sub_sa = "₀₁₂₃₄₅₆₇₈₉"
  res = x.maketrans(''.join(normal), ''.join(sub_sa))
  return x.translate(res)

print('2(C{}H{}) + 9(O{}) = 8(CO{}) + 2(OH{})'.format(get_sub('4'),get_sub('2'),get_sub('2'),get_sub('2'),get_sub('2')))
```
